## **Stage 2 â€“ Data Cleaning and Pre-processing**

## **Data Cleaning**

```python
print(df["Item Fat Content"].unique())
```

```python
df["Item Fat Content"] = df["Item Fat Content"].replace({"LF":"Low Fat",
                                                        "low fat":"Low Fat",
                                                         "reg":"Regular"})
```

## **Handle Missing Data**

```python
df.fillna(df["Item Weight"].mean(),inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

# **Business Requirements**

## **KPI's Requirements**

```python
#total sales
total_sales =df["Sales"].sum()

#average sales
Avg_sales = df["Sales"].mean()

#noof items sold
no_of_items_sold = df["Sales"].count()

#average Ratings
avg_ratings = df["Rating"].mean()

# Display

print(f"Total sales :${total_sales:,.0f}")
print(f"Average sales :${Avg_sales:,.1f}")
print(f"no of items sold:{no_of_items_sold:,.0f}")
print(f"Average Ratings:{avg_ratings:,.1f}")
```

## **Stage 3 â€“ EDA and Visualizations**

### **Total sales by countent**

```python
sales_by_fat =df.groupby("Item Fat Content")['Sales'].sum()


plt.pie(sales_by_fat, labels= sales_by_fat.index ,autopct='%.1f%%', startangle=90)

plt.title("Sales by fat content")
plt.show()
```

Description of the Chart

The chart is a pie chart titled "Sales by fat content".
It shows the distribution of total sales based on Item Fat Content with two categories:

Low Fat â€“ This segment is the largest, making up 64.6% of total sales.

Regular â€“ This segment accounts for 35.4% of total sales.


The Low Fat portion is shown in blue, and the Regular portion is shown in orange.
This indicates that products with Low Fat content contribute nearly two-thirds of the total sales, while Regular fat products contribute about one-third.

## **Total sales by item type**

```python
sales_by_type = df.groupby("Item Type")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(x=sales_by_type.index , height=sales_by_type.values)
plt.title("Total sales by item type")
plt.xticks(rotation=-90)
plt.xlabel("Items type")
plt.ylabel("Total sales")
plt.tight_layout()
plt.show()
```

Description of the Chart

The chart is a horizontal bar chart titled "Total sales by Item Type".
It displays total sales across different product categories in the dataset.

Each bar represents an item type, and its length shows the total sales amount.

Key Observations

Fruits & Vegetables have the highest total sales, indicating they are the most purchased category.

Snack Foods also show strong sales and are among the top-performing categories.

Frozen Foods, Dairy, and Household items also contribute significantly to sales.

Categories like Seafood, Breakfast, and Starchy Foods have relatively lower sales.

Other moderate categories include Canned, Baking Goods, Health & Hygiene, Hard Drinks, and Soft Drinks.


Insight Summary

The chart clearly shows that customers spend most on Fruits & Vegetables and Snack Foods, while categories such as Seafood and Breakfast items see the least sales. This insight can help businesses focus marketing and stocking efforts on high-performing categories.

## **Fat content by outlet for total sales**

```python
grouped = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]


grouped.plot(kind="bar",figsize=(8,5))
plt.title("Outlet Tier by item Fat content")
plt.xlabel("Out location tier")
plt.ylabel("Total sales")
plt.legend(title="items fat content")
plt.tight_layout
plt.show()
```

Description of the Chart

The chart is a grouped bar chart titled "Outlet Tier by Item Fat content".
It compares total sales across different Outlet Tiers (Tier 1, Tier 2, and Tier 3) for two categories of Item Fat Content:

Regular

Low Fat


Each outlet tier has two bars:

Blue represents Regular fat items.

Orange represents Low Fat items.


Key Observations

1. In all outlet tiers, Low Fat items generate higher sales than Regular Fat items.

Low Fat consistently shows longer bars across Tier 1, Tier 2, and Tier 3.

Regular Fat items have lower total sales comparatively.


2. Tier 3 outlets have the highest sales overall

Both Low Fat and Regular categories show their strongest performance in Tier 3, especially Low Fat.


3. Tier 1 has the lowest sales

Tier 1 shows the smallest bars for both categories, indicating the lowest sales volumes.


Overall Insight

Low Fat items are more popular across all outlet types, and

Tier 3 outlets drive the largest portion of total sales.

## **Total sales by outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(8,5))
plt.plot(sales_by_year.index,sales_by_year.values,marker="o",linestyle='-')
plt.title("outlet establishment")
plt.xlabel("out establishment year")
plt.ylabel("Total sales")
plt.tight_layout()
plt.show()
```

Description of the Chart

The chart is a line chart titled "Outlet establishment (year)", showing Total Sales across different Outlet Establishment Years.

Each point on the line represents the sum of sales for outlets established in a particular year.

Key Observations

1. Sales fluctuate across different establishment years.

Some years have noticeably higher sales, while others show lower performance.



2. A peak is visible around one of the mid-range years, where total sales reach the highest point on the chart.


3. Certain years show a sharp decline, indicating older or newer outlets may have weaker performance.


4. Overall Trend:

The sales do not follow a steady upward or downward trend.

The pattern is zig-zag, suggesting that outlet sales depend more on other factors (like location or customer base) rather than the establishment year alone.




Insight Summary

The best-performing establishment year shows strong sales, while

Some specific years show significantly lower sales, indicating variability in outlet performance.

This implies that establishment year is not a consistent predictor of sales.

## **Sales by outlet size**

```python
sales_by_size = df.groupby("Outlet Size")['Sales'].sum()

plt.figure(figsize=(8,5))
plt.pie(sales_by_size,labels=sales_by_size.index,autopct='%1.1f%%',startangle=90)
plt.title("outlet size")
plt.tight_layout()
plt.show()
```

Description of th chart

The chart shown is a pie chart titled "outlet size," which visually represents the proportion of sales from different outlet sizesâ€”Small, Medium, and High. The percentage breakdown is as follows: Small outlets contribute the largest share at 42.3%, Medium outlets contribute 37.0%, and High outlets contribute 20.7%. The chart uses distinct colors for each outlet size and displays the percentage for each category within its respective segment. This kind of visualization helps to quickly understand which outlet size is responsible for the highest sales and allows for easy comparison among the three outlet size categories.



## **Sales by outlet Location**

```python
sales_by_location = df.groupby("Outlet Location Type")['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values("Sales",ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x="Sales",y="Outlet Location Type",data=sales_by_location)
plt.title("Total sales by outlet location Type")
plt.xlabel("Total sales")
plt.ylabel("out location Type")
plt.tight_layout()
plt.show()
```

Describe of the chart

The chart shown is a horizontal bar plot that illustrates the total sales by outlet location type. The y-axis represents different outlet location types, with "Tier 2" and "Tier 3" being displayed, while the x-axis represents the total sales for each location type.From the chart, it is clear that "Tier 3" outlets have higher total sales compared to "Tier 2" outlets. The chart is titled "Total sales by outlet location Type" and provides a clear visual comparison between the two outlet types based on their aggregated sales. The data visualization was created using seaborn's barplot function in Python, with the sales values sorted in descending order, showing the most successful outlet type at the top
